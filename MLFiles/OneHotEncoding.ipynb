```python
import pandas as pd
from sklearn import linear_model
df = pd.read_csv("C:\\Users\\mansi\\Dropbox\\PC\\Downloads\\homeprices.csv")
df
```

```python
dummis = pd.get_dummies(df[['town']])
dummis
```

```python
merged = pd.concat([df,dummis],axis='columns')
merged
```

```python
final = merged.drop(['town','town_west windsor'],axis = 'columns')
final
```

```python
reg = linear_model.LinearRegression()
```

```python
X = final[['area','town_monroe township','town_robinsville']]
y = final[['price']]
X
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2800,0,1]])
```

```python
reg.predict([[3400,0,0]])
```

```python
reg.score(X,y)
```
